```python
%matplotlib nbagg

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = pd.read_csv('Logs/test.log').as_matrix()
Y = pd.read_csv('Logs/train.log').as_matrix()
```

```python
plt.close()
plt.figure(figsize=(12,8))
plt.plot(np.arange(0, X.shape[0]), X, label='Test Accuracy')
plt.plot(np.arange(0, Y.shape[0]), Y, label='Train Accuracy')
plt.title(r'Accuracy vs Epoch for best performing model (Noise, Dropout, Convolution & Rotation)')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='best')

#plt.savefig('bestPerformance.png')
```
